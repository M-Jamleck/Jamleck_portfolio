```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('/content/malaria_infection.csv', encoding='ISO-8859-1')
df.head(17)
```

```python
# Remove duplicates
df_cleaned = df.drop_duplicates()
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Data type information and non-null counts
df.info()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# check for remaining null values
df.isnull().sum()
```

```python
# Convert columns to numeric types
df['No. of cases_median'] = pd.to_numeric(df['No. of cases_median'], errors='coerce')
df['No. of deaths_median'] = pd.to_numeric(df['No. of deaths_median'], errors='coerce')

# Convert WHO Region to categorical data
df['WHO Region'] = df['WHO Region'].astype('category')
```

```python
# Create a new column for case-fatality ratio
df['case_fatality_ratio'] = df['No. of deaths_median'] / df['No. of cases_median']
df.head()
```

```python
# Apply log transformation to cases and deaths
df['log_cases'] = df['No. of cases_median'].apply(lambda x: np.log(x + 1))
print(df[['No. of cases_median', 'log_cases']])
```

```python
# Data Visualization
sns.boxplot(x='WHO Region', y='No. of cases_median', data=df)
plt.show()
```

```python
# line plot
plt.plot(df['Year'], df['No. of deaths'], marker='o')
plt.title('Number of Deaths Over the Years')
plt.xlabel('Year')
plt.ylabel('No. of Deaths')
plt.grid(True)
plt.show()
```

```python
# pair plot visualizing relationships between multiple variables in a dataset
columns_to_plot = ['No. of cases_median', 'No. of deaths_median', 'No. of cases_min']
sns.pairplot(df[columns_to_plot])
plt.show()
```

```python
from scipy.stats import pearsonr

# Extract the relevant columns
cases = df['No. of cases_median']
deaths = df['No. of deaths_median']

# Perform Pearson's correlation test
corr, p_value = pearsonr(cases, deaths)

print(f"Pearson correlation coefficient: {corr}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant relationship between cases and deaths.")
else:
    print("Fail to reject the null hypothesis. There is no significant relationship between cases and deaths.")
```

```python
# Export Data
df_export = df[['Country','Year','No. of cases','No. of deaths','No. of cases_median', 'No. of cases_min','No. of cases_max','No. of deaths_median','No. of deaths_min','No. of deaths_max', 'WHO Region']]
df_export.to_csv('/content/malaria_infection.csv', index=False)
```
